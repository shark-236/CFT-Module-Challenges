```python
import csv
from pathlib import Path
#Part 1: loan_costs is a given list. number_of_loans is created to define the length of the list using the 'len' function.
# loans_total is created to defined the total cost using the 'sum' function. average_loan_price is created to calculate the
# average of all the loans. The defined variables are then printed. 
loan_costs = [500, 600, 200, 1000, 450]
number_of_loans = len(loan_costs)
print("The total number of loans is", number_of_loans)
loans_total = sum(loan_costs)
print("The total value of the loans is", loans_total)
average_loan_price = loans_total / number_of_loans
print("The average loan price is", average_loan_price)

#Part 2: loan is a given dictionary. future_value and remaining_months are created to call in the related values from the dictionary.
# discount_rate is created to call in the given discount rate. present_value is created to calculate a fair value of a loan.
# The if/else conditional is created to evaluate the viability ot purchasing the loan.
loan = {
    "loan_price": 500,
    "remaining_months": 9,
    "repayment_interval": "bullet",
    "future_value": 1000,
}
future_value = loan.get("future_value")
remaining_months = loan.get("remaining_months")
print("The future value of this loan is $", future_value, "and there are", remaining_months, "months remaining on this loan.")
discount_rate = 0.20
present_value = future_value / ((1 + discount_rate/12)**remaining_months)
print("The fair value of this loan is", present_value)
loan_price = loan.get("loan_price")
if present_value >= loan_price:
    print("This loan is worth at least the cost to buy it.")
else:
    print("The loan is too expensive and not worth the cost.")

#Part 3: new_loan is a given dictionary. The function calculate_present_value is created to automate the calculation given the relevant
# data points. "global" is used to call in a global variable, and return the updated value from the local function. The resulting
# value is then printed.
new_loan = {
    "loan_price": 800,
    "remaining_months": 12,
    "repayment_interval": "bullet",
    "future_value": 1000,
}
def calculate_present_value(future_value, annual_discount_rate, remaining_months):
    global present_value
    present_value = future_value / ((1 + annual_discount_rate/12)**remaining_months)
    return present_value
calculate_present_value(1000, 0.2, 12)
print("The present value of the new loan is", present_value)

#Part 4: loans is a given dictionary. inexpensive_loans is created as an empty list. A 'for' loop is then created to cycle through
# the values in the dictionary and add loans less than or equal to 500 into inexpensive_loans. The resulting list is then printed. 
loans = [
    {
        "loan_price": 700,
        "remaining_months": 9,
        "repayment_interval": "monthly",
        "future_value": 1000,
    },
    {
        "loan_price": 500,
        "remaining_months": 13,
        "repayment_interval": "bullet",
        "future_value": 1000,
    },
    {
        "loan_price": 200,
        "remaining_months": 16,
        "repayment_interval": "bullet",
        "future_value": 1000,
    },
    {
        "loan_price": 900,
        "remaining_months": 16,
        "repayment_interval": "bullet",
        "future_value": 1000,
    },
]
inexpensive_loans = [] 
for item in loans:
    loan_price = item["loan_price"]
    if loan_price <= 500:
        inexpensive_loans.append(item)
print(inexpensive_loans)

#Part 5: header is given. The code following exists to write the code into a new csvfile, add the header, and add the loans
# previously identified through the 'for' loop for inexpensive_loans
header = ["loan_price", "remaining_months", "repayment_interval", "future_value"]
csvpath = Path("inexpensive_loans.csv")
with open(csvpath, "w") as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=",")
    csvwriter.writerow(header)
    for item in inexpensive_loans:
        csvwriter.writerow(item.values())
```
